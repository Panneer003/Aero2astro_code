```python
# Load images from Dataset folder
import os
import cv2
img_path = '/content/gdrive/MyDrive/yolov4_custom_data/Optimization/tensorflow-yolov4-tflite/Dataset'
save_path = '/content/gdrive/MyDrive/PP-Yolo_custom/PaddleDetection/dataset/coco/images'
i =1
for file in os.listdir(img_path):
  if file.endswith('.jpg'):
    from_path = os.path.join(img_path,file)
    to_path = os.path.join(save_path,file)
    img = cv2.imread(from_path)
    cv2.imwrite(to_path,img)
    print('copied {}'.format(i))
    i +=1
    
```
