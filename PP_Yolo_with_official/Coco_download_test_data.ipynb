```python
from pycocotools.coco import COCO
import requests

# instantiate COCO specifying the annotations json path
coco = COCO('/content/annotations/instances_val2014.json')
# Specify a list of category names of interest
catIds = coco.getCatIds(catNms=['bird'])
# Get the corresponding image ids and images using loadImgs
imgIds = coco.getImgIds(catIds=catIds)
images = coco.loadImgs(imgIds)

annIds = coco.getAnnIds(catIds = catIds,iscrowd=0)
annot = coco.loadAnns(annIds)
```

```python
# Download coco datset from 1000 instaces from 1100 instaces
# Save the images into a local folder and create annotations dictionary 
ann_file = {}
ann_file['images'] = []
ann_file['annotations'] = []
i = 0
count = 0

for im in images:
  download = False
  for an in annot:
      if an['image_id'] == im['id']:
        count +=1
        if count >=1000:
          ann_file['annotations'].append(an)
          i +=1
          print('{} handled..'.format(i))
          if i==100:
            break
          if download == False:
            img_data = requests.get(im['coco_url']).content
            with open('bird/' + im['file_name'], 'wb') as handler:
              handler.write(img_data)
            ann_file['images'].append(im)
            download = True

  if i==100 :
      break  
      
  
```

```python
# Create json file
import json 

with open('bird.json','w') as f:
  json.dump(ann_file,f)
```
